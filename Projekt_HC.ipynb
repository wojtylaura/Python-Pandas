Analiza danych WIG na rok 2023.
Wczytany plik przechowuje dane od 01-01-2023 do 31-12-2023.

```python
import pandas as pd

file = 'wig_d.csv'
data = pd.read_csv(file)

print(data)
```

Na samym początku, żeby ograniczyć nasz zestaw danych wybierzemy tylko te wiersze, dla których kurs otwarcia był wyższy od 52000 i mniejszy od 63000

```python
data = data[(data['Otwarcie'] >= 58000) & (data['Otwarcie'] <= 63000)]
print(data)
```

Sortujemy dane rosnąco względem najwyższej wartości 

```python
data = data.sort_values(by = 'Najwyzszy')
print(data)
```

Daodajemy nową zmienną - procentową zmianę między kursem otwarcia, a kursem zamknięcia

```python
data['Zmiana_proc'] = (data['Zamkniecie'] - data['Otwarcie'])/data['Otwarcie']
print(data[['Zmiana_proc']])
```

```python
print(data)
```

Wykonujemy kilka podstawowych obliczeń z biblioteki 'statistics'

```python
import statistics 

mean_open = round(statistics.mean(data['Otwarcie']),2)
median_close = statistics.median(data['Zamkniecie'])
mode_higest = statistics.mode(data['Najwyzszy'])
max_vol = max(data['Wolumen'])
min_vol = min(data['Wolumen'])
dev_zm_proc = statistics.pstdev(data['Zmiana_proc'])

print(mean_open, median_close, mode_higest, max_vol, min_vol, dev_zm_proc)
```

Wykres średnich wartości otwarcia dla poszczególnych miesięcy 

```python
import pandas as pd
import matplotlib.pyplot as plt

data['Data'] = pd.to_datetime(data['Data'])
data['Miesiąc'] = data['Data'].dt.to_period('M')
data_grouped = data.groupby('Miesiąc').mean()
 
plt.figure(figsize=(10, 7))
plt.plot(data_grouped.index.astype(str), data_grouped['Otwarcie'], 'm.-')
plt.xticks(rotation = 45)
plt.yticks(data_grouped['Otwarcie'], fontsize = 10)
plt.xlabel('Miesiąc')
plt.ylabel('Otwarcie')
plt.title('Średnie wartości otwarcia w poszczególnych miesiącach')

plt.show()
```

Porównanie średniej najniższej wartości ze średnią najwyższą wartością dla każdego miesiąca

```python
data_grouped['Różnica'] = data_grouped['Najwyzszy'] - data_grouped['Najnizszy']

plt.plot(data_grouped.index.astype(str), data_grouped['Najwyzszy'], 'm.-', label = 'Najwyższy')
plt.plot(data_grouped.index.astype(str), data_grouped['Najnizszy'], 'b.-', label = 'Najniższy')

for i, row in data_grouped.iterrows():
    month_str = i.strftime('%Y-%m')
    plt.vlines(x = month_str, ymin = row['Najnizszy'], ymax = row['Najwyzszy'], color = 'black', linestyle = '--')
    point = (row['Najwyzszy'] + row['Najnizszy']) / 2
    plt.text(month_str, point, f'{row["Różnica"]:.2f}', ha = 'center', va = 'center', fontsize = 9, color = 'black')
    
plt.xticks(rotation = 45)
plt.xlabel('Miesiąc')
plt.ylabel('Wartość średnia')
plt.title('Porównanie średnich najwyższych i najniższych wartości dla każdego miesiąca')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.plot(data_grouped.index.astype(str),data_grouped['Zmiana_proc'],'.--', color = '#FF9930' )
plt.yticks(data_grouped['Zmiana_proc'])
plt.title('Zmiana procentowa dla każdego miesiąca')
plt.xlabel('Miesiąc')
plt.ylabel('Zmiana procentowa')
plt.grid()
plt.show()
```
